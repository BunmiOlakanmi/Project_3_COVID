```python
#import dependencies
import os
import csv
import pandas as pd
```

```python
# Path to collect data from the Resources folder

#1 - Death Counts for Infections - Influenza, Pnuemonia, COVID-19

file_path = os.path.join('..', 'References', 'Death_Counts_for_infections_Influenza__Pneumonia__and_COVID-19.csv')

# Import the Death_Counts_for_infections_Influenza__Pneumonia__and_COVID-19.csv file as a DataFrame
death_counts_df=pd.read_csv(file_path)

death_counts_df.head(5)
```

```python
# Finding how many unique Jurisdictions are provided
death_counts_df["Jurisdiction"].unique() ##value_counts()

####  HHS Regions: https://www.hhs.gov/about/agencies/iea/regional-offices/index.html
### exclude since we cannot aggregate with other datasources
```

```python
##remove HHS Regions from data
clean_df = death_counts_df[~death_counts_df["Jurisdiction"].str.contains('HHS Region')] 
clean_df["Jurisdiction"].unique()
```

```python
clean_df.describe()
```

```python
### Remove the All Ages category from the "Age Group" Column
clean_df2=clean_df[clean_df["Age Group"]!="All Ages"]

### Remove the United States category from the "Jurisdiction" Column
clean_df2=clean_df2[clean_df2["Jurisdiction"]!="United States"]
clean_df2
```

```python
#Group by Jurisdiction and Age Group - mean of remaining columns
age_grp1_df=clean_df2.groupby(['Jurisdiction', 'Age Group']).mean()
age_grp1_df
age_grp1_df=age_grp1_df.rename(
    columns= {'COVID-19 Deaths':"Average COVID-19 Deaths",
              'Total Deaths':"Average Total Deaths",
              "Pneumonia Deaths":"Average Pneumonia Deaths",
              "Influenza Deaths":"Average Influenza Deaths",
              "Pneumonia or Influenza":"Average Pneumonia or Influenza",
              "Pneumonia, Influenza, or COVID-19 Deaths":"Average Pneumonia, Influenza, or COVID-19 Deaths"         
            })
age_grp1_df.head(5)




```

```python
#Group by Jurisdiction and Age Group - sum of remaining columns
age_grp2_df=clean_df2.groupby(['Jurisdiction', 'Age Group']).sum()
age_grp2_df.head(5)
```

```python
#Group by Jurisdiction  - mean of remaining columns
age_grp3_df=clean_df2.groupby(['Jurisdiction']).mean()
age_grp3_df=age_grp3_df.rename(
    columns= {'COVID-19 Deaths':"Average COVID-19 Deaths",
              'Total Deaths':"Average Total Deaths",
              "Pneumonia Deaths":"Average Pneumonia Deaths",
              "Influenza Deaths":"Average Influenza Deaths",
              "Pneumonia or Influenza":"Average Pneumonia or Influenza",
              "Pneumonia, Influenza, or COVID-19 Deaths":"Average Pneumonia, Influenza, or COVID-19 Deaths"         
            })
age_grp3_df.head(5)

```

```python
#Group by Jurisdiction  - sum of remaining columns
age_grp4_df=clean_df2.groupby(['Jurisdiction']).sum()
age_grp4_df.head(5)
```

```python
#Group by Age Group  - mean of remaining columns
age_grp5_df=clean_df2.groupby(['Jurisdiction']).mean()
age_grp5_df=age_grp5_df.rename(
    columns= {'COVID-19 Deaths':"Average COVID-19 Deaths",
              'Total Deaths':"Average Total Deaths",
              "Pneumonia Deaths":"Average Pneumonia Deaths",
              "Influenza Deaths":"Average Influenza Deaths",
              "Pneumonia or Influenza":"Average Pneumonia or Influenza",
              "Pneumonia, Influenza, or COVID-19 Deaths":"Average Pneumonia, Influenza, or COVID-19 Deaths"         
            })
age_grp5_df.head(5)
```

```python

#Group by Age Group  - sum of remaining columns
age_grp6_df=clean_df2.groupby(['Jurisdiction']).sum()
age_grp6_df.tail(5)
```

```python
Influenza_table1 = age_grp6_df[["COVID-19 Deaths","Total Deaths","Pneumonia or Influenza"]]
Influenza_table1 = Influenza_table1.rename(columns={"COVID-19 Deaths":"COVID-19_Deaths","Total Deaths": "Total_Deaths","Pneumonia or Influenza": "Pneumonia_Influenza_Deaths"})
Influenza_table1
```

```python
###### Collect lat and long
#1 - US Cities Longitudes and Latitudes

file_path2 = os.path.join('..', 'References', 'US Cities Long_Lat.csv')

# Import the US Cities Long_Lat.csv file as a DataFrame
uslat_lon_df=pd.read_csv(file_path2)

uslat_lon_df.head()

```

```python
####Append Longitude and Latitude to the Influenza_table
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

## df.reset_index(level=['tick', 'obs'])

Influenza_table2= Influenza_table1.reset_index(level='Jurisdiction', inplace=False)

Influenza_table2 = Influenza_table2.merge(uslat_lon_df, how='inner', 
                                        left_on=["Jurisdiction"], 
                                        right_on =["NAME"],
                                        suffixes=('', '_y')
                                       )

Influenza_table =Influenza_table2.set_index('Jurisdiction', inplace=False)
Influenza_table 
```
